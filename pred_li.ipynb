```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
#poly_reg=PolynomialFeatures(degree=4)
#indep_poly=poly_reg.fit_transform(indep)
from sklearn.pipeline import Pipeline #pipelining  
```

```python
import sklearn.model_selection as model_selection
```

```python
years1=wp.columns
years1=years1[:-1]
years1
```

```python
dep=countries_transpose['India'] # dependent variables
dep.values
```

```python
y_reg=np.array(dep)
x_reg=np.array(pd.to_numeric(years1)).reshape((-1,1))
print(x_reg)
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(x_reg, y_reg, train_size=0.65,test_size=0.35, random_state=101)
```

```python
print ("X_train: ", X_train)
print ("y_train: ", y_train)
print("X_test: ", X_test)
print ("y_test: ", y_test)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)

```

```python
r_sq = model.score(X_test, y_test)
print('coefficient of determination:', r_sq)
print('intercept:', model.intercept_)
print('slope:', model.coef_)
```

```python
x_predict=np.array([2014,2015,2016,2017,2018,2019,2020,2021,2022,2023,2024,2025,2026,2027,2028,2029,2030]).reshape(-1,1)
y_pred = model.predict(x_predict)
print('predicted response:', y_pred, sep='\n')
```

```python
%matplotlib inline
plt.xlabel('Year',fontsize=20)
plt.ylabel('Population',fontsize=20)
plt.plot(x_predict,y_pred,color='red',marker='+')
```
